### Creating Term-Frequency Vectorization Model For Search
This will be used to create a method of searching for anime.<br> First we need to clean the name of the title of all special characters to make it easier for the user to search for the anime.


```python
# Cleaning title of all ancillary characters not normally used. 
import re

# Method for removing special characters
def clean_name(name):
    name = re.sub("[^a-zA-Z0-9 ]", "", name)
    return name

# Adds column to dataframe with cleaned title name
anime_df["cleaned_name"] = anime_df["name"].apply(clean_name)

display(anime_df)
```

We will then use TfidVectorizer and cosined_similarity from sklearn library to create a responsive search function. If not complete matches are found, then fuzzy matching will occur to match partial matches. 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from fuzzywuzzy import process

# Create and fit the TF-IDF vectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf = vectorizer.fit_transform(anime_df["cleaned_name"])

def search(query, top_n=10, similarity_threshold=0.1, fuzzy_cutoff=60):
    """
    Search for anime titles using TF-IDF, cosine similarity, and fuzzy matching
    """
    cleaned_query = clean_name(query)
    query_vector = vectorizer.transform([cleaned_query])
    similarity = cosine_similarity(query_vector, tfidf).flatten()
    
    # Get indices of results above the similarity threshold
    relevant_indices = similarity > similarity_threshold
    
    if not any(relevant_indices):
        # If no relevant results, use fuzzy matching on all titles
        fuzzy_matches = process.extractBests(query, anime_df['name'], score_cutoff=fuzzy_cutoff, limit=top_n)
        if fuzzy_matches:
            fuzzy_df = pd.DataFrame(fuzzy_matches, columns=['name', 'fuzzy_score', 'index'])
            final_results = anime_df.loc[fuzzy_df['index']].copy()
            final_results['combined_score'] = fuzzy_df['fuzzy_score'] / 100
        else:
            return pd.DataFrame()  # Return empty DataFrame if no matches
    else:
        # Create a DataFrame with initial results
        results = anime_df[relevant_indices].copy()
        results['similarity_score'] = similarity[relevant_indices]
        
        # Perform fuzzy matching on these results
        fuzzy_matches = process.extractBests(query, results['name'], score_cutoff=fuzzy_cutoff, limit=top_n)
        
        # Create a new DataFrame with fuzzy match results
        fuzzy_df = pd.DataFrame(fuzzy_matches, columns=['name', 'fuzzy_score', 'index'])
        
        # Merge fuzzy results with original results
        final_results = results.merge(fuzzy_df, on='name', how='inner')
        
        # Calculate a combined score (you can adjust the weights)
        final_results['combined_score'] = (final_results['similarity_score'] * 0.5) + (final_results['fuzzy_score'] / 100 * 0.5)
    
    # Sort by combined score and get top N
    final_results = final_results.sort_values('combined_score', ascending=False).head(top_n)
    
    return final_results[['anime_id', 'name', 'genre', 'type', 'combined_score']]


# Function to get the best match
def get_best_match(query):
    results = search(query, top_n=1)
    if not results.empty:
        return results.iloc[0]['name']
    return None
```

Here is the initial testing of the search function. When the submit button is clicked, the anime that is typed will be search for. This will return the top 10 closest names. 

```python
import ipywidgets as widgets

# Creates a text box for the user to input text
anime_input = widgets.Text(
    value="Pokemon",
    description="Anime Name:",
    disabled=False
)

# Create a submit button
submit_button = widgets.Button(
    description = "Search",
    button_style = 'primary',
    tooltip = 'Click the search'
)

# Output widgets used to display results
anime_list = widgets.Output()

# Function to display names that are most similar while typing >3 characters
# Function to handle button click
def on_button_click(b):
    with anime_list:
        anime_list.clear_output()
        name = anime_input.value
        if len(name) > 3:
            results = search(name)
            if not results.empty:
                print(f"Search results for '{name}':")
                display(results)
            else:
                print(f"No results found for '{name}'. Please try a different search term.")
        else:
            print("Please enter at least 4 characters.")

# Assign the function to the button
submit_button.on_click(on_button_click)

# Create a horizontal box to hold the input and button side by side
input_area = widgets.HBox([anime_input, submit_button])

# Display the widgets
display(input_area, anime_list)
```

### Preparing Data
The two data frames are taken an put into a user-anime matrix.<br> Here you can see the matrix displayed below. NaN values are replaced with -1 to ensure that no errors occur when parsing the matrix for recommendations.

```python
"""This uses a lot of memory, so please be patient"""
# Get average value of duplicate ratings
rating_df = rating_df.groupby(['user_id', 'anime_id'], as_index=False)['rating'].mean()

# Create user-item matrix
def prepare_data(rating_df, anime_df):
    user_anime_matrix = rating_df.pivot(index='user_id', columns='anime_id', values='rating').fillna(-1)
    return user_anime_matrix

user_anime_matrix = prepare_data(rating_df, anime_df)

# Convert to numpy array
X = user_anime_matrix.values

# Check the shape of X
print("Shape of (user-anime matrix):", X.shape)

display(user_anime_matrix)
```

### Collaborative Filtering Implementation
This is where collaborative filtering is going to be implmented. Prior to doing this, we need to make sure that the data is consistent throughout all the dataframes. If it is not, then a method for handling such a case must be provided. <br> Here you can see that we are missing anime 1735 from both the rating_df and user_anime_matrix. In our case, if the item is not found in the matrix, then popular animes from the same genre will instead be recommended to the user. 

```python
# This function checks to see if a certain anime is in the user-item matrix. If it is not present, then a try block needs implemented to prevent errors
def check_data_consistency(anime_name, anime_df, rating_df, user_anime_matrix):
    print(f"Checking data consistency for '{anime_name}':")
    
    # Check in anime_df
    anime_info = anime_df[anime_df['name'] == anime_name]
    if anime_info.empty:
        print("  Not found in anime_df")
    else:
        anime_id = anime_info.iloc[0]['anime_id']
        print(f"  Found in anime_df with ID: {anime_id}")
    
    # Check in rating_df
    ratings = rating_df[rating_df['anime_id'] == anime_id]
    if ratings.empty:
        print("  No ratings found in rating_df")
    else:
        print(f"  Found {len(ratings)} ratings in rating_df")
    
    # Check in user_anime_matrix
    if anime_id in user_anime_matrix.columns:
        print("  Found in user_anime_matrix")
    else:
        print("  Not found in user_anime_matrix")

# Checking to see if anime is in the user-item matrix
check_data_consistency("Naruto: Shippuuden", anime_df, rating_df, user_anime_matrix)
```

```python
import pandas as pd
import numpy as np

# Find all the users that have rated the anime 8 or higher
def get_similar_users(user_anime_matrix, anime_id, min_rating=8):
    if anime_id not in user_anime_matrix.columns:
        return []
    anime_ratings = user_anime_matrix[anime_id]
    similar_users = anime_ratings[anime_ratings >= min_rating].index.tolist()
    return similar_users
```

```python
def get_recommendations(anime_title, anime_df, rating_df, user_anime_matrix, min_rating=8, similarity_threshold=0.1):
    """Recommendation Function
 This funcion takes all the ratings that similar users also liked, but only the top the animes that >10% also liked.
It then calulcates what percent of ALL users also liked the content. The ratio score is then calculated using (similar_user_percent / all_user_percent). 
The wider the margin, the greater the score. Recommendations are then provided descending from highest ratio score.
 """
    try:
        anime_info = anime_df[anime_df['name'] == anime_title]
        if anime_info.empty:
            print(f"Anime '{anime_title}' not found in the database.")
            return [], [], [], []
        anime_id = anime_info.iloc[0]['anime_id']
    except Exception as e:
        print(f"Error finding anime '{anime_title}': {str(e)}")
        return [], [], [], []

    if anime_id not in rating_df['anime_id'].values:
        print(f"No ratings found for anime '{anime_title}' (ID: {anime_id}).")
        return [], [], [], []

    if anime_id not in user_anime_matrix.columns:
        print(f"Anime '{anime_title}' (ID: {anime_id}) not in user-anime matrix. Rebuilding matrix...")
        user_anime_matrix = rating_df.pivot(index='user_id', columns='anime_id', values='rating')

    similar_users = get_similar_users(user_anime_matrix, anime_id, min_rating)
    
    if not similar_users:
        print(f"No users found who rated '{anime_title}' {min_rating} or higher.")
        return [], [], [], []
    
    similar_users_ratings = rating_df[rating_df['user_id'].isin(similar_users) & (rating_df['rating'] >= min_rating)]
    
    anime_counts = similar_users_ratings['anime_id'].value_counts()
    anime_percentages = anime_counts / len(similar_users)
    
    recommended_anime_ids = anime_percentages[anime_percentages >= similarity_threshold].index.tolist()
    
    all_high_ratings = rating_df[rating_df['rating'] >= min_rating]
    overall_percentages = all_high_ratings['anime_id'].value_counts() / rating_df['user_id'].nunique()
    
    recommendations = []
    for anime_id in recommended_anime_ids:
        if anime_id in overall_percentages.index:
            similar_users_pct = anime_percentages[anime_id]
            all_users_pct = overall_percentages[anime_id]
            ratio_score = similar_users_pct / all_users_pct if all_users_pct > 0 else 1000 
            recommendations.append((anime_id, similar_users_pct, all_users_pct, ratio_score))
    
    recommendations.sort(key=lambda x: x[3], reverse=True)
    
    recommended_anime = []
    similar_user_scores = []
    all_user_scores = []
    ratio_scores = []
    
    for anime_id, similar_pct, all_pct, ratio in recommendations:
        name = anime_df[anime_df['anime_id'] == anime_id]['name'].values[0]
        recommended_anime.append(name)
        similar_user_scores.append(similar_pct)
        all_user_scores.append(all_pct)
        ratio_scores.append(ratio)
    
    return recommended_anime, similar_user_scores, all_user_scores, ratio_scores

```

```python
def get_popular_genre_recommendations(anime_title, anime_df, rating_df, n=10):
    """Genre Recommendation Function to handle the case where searched anime has zero(0) ratings"""
    anime_info = anime_df[anime_df['name'] == anime_title].iloc[0]
    genre = anime_info['genre']
    
    # Get all animes in the same genre
    same_genre_animes = anime_df[anime_df['genre'].str.contains(genre, na=False)]
    
    # Calculate popularity (number of ratings) for each anime
    popularity = rating_df['anime_id'].value_counts().reset_index()
    popularity.columns = ['anime_id', 'num_ratings']
    
    # Merge popularity with anime info
    popular_genre_animes = same_genre_animes.merge(popularity, on='anime_id', how='left')
    popular_genre_animes['num_ratings'] = popular_genre_animes['num_ratings'].fillna(0)
    popular_genre_animes = popular_genre_animes.sort_values('num_ratings', ascending=False)
    
    # Return top N recommendations
    recommendations = popular_genre_animes.head(n)
    return recommendations[['name', 'genre', 'num_ratings']]
```

```python
def get_recommendations_wrapper(anime_name, anime_df, rating_df, user_anime_matrix):
    """Wrapper Function to handle user only providing Anime Name"""
    # Get top 5 search results
    search_results = search(anime_name, top_n=5)  
    if search_results.empty:
        return pd.DataFrame(), pd.DataFrame()

    best_match = get_best_match(anime_name)
    
    recommended_anime, similar_user_scores, all_user_scores, ratio_scores = get_recommendations(
        best_match, anime_df, rating_df, user_anime_matrix
    )
    
    if not recommended_anime:
        print(f"No recommendations found based on user ratings. Suggesting popular animes in the same genre.")
        recommendations_df = get_popular_genre_recommendations(best_match, anime_df, rating_df)
    else:
        recommendations_df = pd.DataFrame({
            'anime_id': [anime_df[anime_df['name'] == anime]['anime_id'].values[0] for anime in recommended_anime],
            'name': recommended_anime,
            'genre': [anime_df[anime_df['name'] == anime]['genre'].values[0] for anime in recommended_anime],
            'similar_users_pct': [score * 100 for score in similar_user_scores],
            'all_users_pct': [score * 100 for score in all_user_scores],
            'ratio_score': ratio_scores
        })
        recommendations_df = recommendations_df.sort_values('ratio_score', ascending=False).head(10).reset_index(drop=True)
    
    return search_results, recommendations_df
```

```python
#Plot a comparison of recommendation scores between similar users and all users
def plot_recommendation_comparison(anime_title, recommendations, similar_user_scores, all_user_scores, ratio_scores, top_n=10):
    plt.figure(figsize=(14, 10))
    x = range(min(len(recommendations), top_n))
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 16))
    
    # Bar plot for percentages
    ax1.bar(x, similar_user_scores[:top_n], alpha=0.8, label='Similar Users')
    ax1.bar(x, all_user_scores[:top_n], alpha=0.5, label='All Users')
    ax1.set_xlabel('Anime')
    ax1.set_ylabel('Percentage of Users Who Liked (Rating >= 8)')
    ax1.set_title(f"Top {top_n} Recommendations Based on '{anime_title}'")
    ax1.set_xticks(x)
    ax1.set_xticklabels(recommendations[:top_n], rotation=45, ha='right')
    ax1.legend()
    
    # Bar plot for ratio scores
    ax2.bar(x, ratio_scores[:top_n], alpha=0.8, color='green')
    ax2.set_xlabel('Anime')
    ax2.set_ylabel('Ratio Score (Similar Users % / All Users %)')
    ax2.set_title(f"Ratio Scores for Top {top_n} Recommendations")
    ax2.set_xticks(x)
    ax2.set_xticklabels(recommendations[:top_n], rotation=45, ha='right')
    
    plt.tight_layout()
    plt.show()

def plot_genre_recommendations(input_anime, recommendations):
    with plot_output:
        clear_output(wait=True)
        plt.figure(figsize=(12, 6))
        sns.barplot(x='name', y='num_ratings', data=recommendations)
        plt.title(f"Popular Anime in the Same Genre as '{input_anime}'")
        plt.xlabel('Anime Title')
        plt.ylabel('Number of Ratings')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
```

### Widget to Enable User To Search For Animes
*If the anime has ratings, then a recommendation list will appear with ratio scores. If not, then popular genres related to the anime will show instead.*
There are a possiblity of three bar plots that appear once a search has finished. If the anime is matched with user input, then two bar plots will appear.<br>
1. A bar plot to show the difference in percentages for similar users and all users that rated the anime >=8
2. A bar plot to show the ratio scores for the top 10 recommended animes. <br><br>
If the searched anime does not appear in the matrix, then popular recommended genres will be listed. This will show only a singular plot.<br>
1. A bar plot to show popular animes within the same genre as the searched animes. Y-axis shows the number of ratings, X-axis shows the anime title.

   * You can try to partially search for animes like typing 'poke' and pokemon should appear. Additionally, you can try to have a non-rated anime appear to show popular genre recommendations. For this, you can try typing "Naruto: Shippuuden"

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import seaborn as sns

# Creates a text box for the user to input text
anime_input = widgets.Text(
    value="Pokemon",
    description="Anime Name:",
    disabled=False
)

# Create a submit button
submit_button = widgets.Button(
    description="Search",
    button_style='primary',
    tooltip='Click to search'
)

anime_list = widgets.Output()
plot_output = widgets.Output()

# Function to display names that are most similar while typing >3 characters
def on_button_click(b):
    with anime_list:
        clear_output(wait=True)
        name = anime_input.value
        if len(name) > 3:
            search_results, recommendations = get_recommendations_wrapper(name, anime_df, rating_df, user_anime_matrix)
            if not search_results.empty:
                print(f"\nSearch Results for '{name}':")
                display(search_results)
                
                if not recommendations.empty:
                    print(f"\nTop 10 recommendations for fans of '{search_results.iloc[0]['name']}':")
                    display(recommendations)
                    
                    # Plot recommendation scores
                    with plot_output:
                        clear_output(wait=True)
                        if 'ratio_score' in recommendations.columns:
                            plot_recommendation_comparison(search_results.iloc[0]['name'],
                                                           recommendations['name'],
                                                           recommendations['similar_users_pct'] / 100,
                                                           recommendations['all_users_pct'] / 100,
                                                           recommendations['ratio_score'])
                        elif 'genre' in recommendations.columns:
                            plot_genre_recommendations(search_results.iloc[0]['name'], recommendations)
                        else:
                            print("Not enough data to create a plot.")
                    
                    # Additional analysis
                    print("\nSummary Statistics:")
                    if 'ratio_score' in recommendations.columns:
                        print(f"Average Ratio Score: {recommendations['ratio_score'].mean():.2f}")
                        print(f"Highest Ratio Score: {recommendations['ratio_score'].max():.2f}")
                        print(f"Lowest Ratio Score: {recommendations['ratio_score'].min():.2f}")
                    elif 'num_ratings' in recommendations.columns:
                        print(f"Average Number of Ratings: {recommendations['num_ratings'].mean():.2f}")
                        print(f"Highest Number of Ratings: {recommendations['num_ratings'].max():.0f}")
                        print(f"Lowest Number of Ratings: {recommendations['num_ratings'].min():.0f}")
                    else:
                        print("No numerical data available for summary statistics.")
                    
                    if 'genre' in recommendations.columns:
                        print("\nGenre Analysis:")
                        genre_counts = recommendations['genre'].str.split(', ', expand=True).stack().value_counts()
                        print(genre_counts)
                    else:
                        print("\nNo genre data available for analysis.")
                else:
                    print(f"\nNo recommendations found for '{search_results.iloc[0]['name']}'.")
            else:
                print(f"No anime found matching '{name}'. Please try a different search term.")
        else:
            print("Please enter at least 4 characters.")

# Assign the function to the button
submit_button.on_click(on_button_click)

# Create a horizontal box to hold the input and button side by side
input_area = widgets.HBox([anime_input, submit_button])

# Display widgets
display(input_area, anime_list, plot_output)
```
